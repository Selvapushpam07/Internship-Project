# Identify peak order items

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
data = {
    'Customer_Name': ['Min-Jun','Ji-Hoon','Seo-Yeon','Ha-Neul','Yu-Jin',
                      'Min-Jun','Ji-Hoon','Seo-Yeon','Ha-Neul','Yu-Jin',
                      'Jae-Hyun','Su-Min','Dong-Hyun','Eun-Ji','Hye-Jin'],

    'Food_Item': ['Kimchi Fried Rice','Bibimbap','Tteokbokki','Bulgogi','Japchae',
                  'Kimchi Fried Rice','Bibimbap','Tteokbokki','Bulgogi','Japchae',
                  'Samgyeopsal','Kimbap','Jajangmyeon','Kimchi Fried Rice','Bibimbap'],

    'Order_Time': ['09:00','10:00','12:00','13:00','18:00',
                   '19:00','20:00','21:00','18:00','19:00',
                   '20:00','12:00','13:00','18:00','19:00'],

    'Location': ['Gangnam','Myeongdong','Hongdae','Itaewon','Gangnam',
                 'Hongdae','Gangnam','Myeongdong','Itaewon','Hongdae',
                 'Gangnam','Myeongdong','Hongdae','Gangnam','Itaewon'],

    'Price': [250,300,200,450,350,250,300,200,450,350,500,150,280,250,300]
}

df = pd.DataFrame(data)

# Peak Order Time Analysis
peak_orders = df.groupby('Order_Time').size().reset_index(name='Orders')

print(peak_orders)

# Line Chart
plt.figure(figsize=(8,5))
plt.plot(peak_orders['Order_Time'], peak_orders['Orders'], marker='o')
plt.title('Peak Order Times')
plt.xlabel('Order Time')
plt.ylabel('Number of Orders')
plt.grid(True)
plt.show()
```

# Finding the most Popular dishes

```python
import pandas as pd
import matplotlib.pyplot as plt

# Find Most Ordered Dishes
dish_count = df['Food_Item'].value_counts()

print("===== MOST POPULAR DISHES =====")
print(dish_count)

# Bar Chart
plt.figure(figsize=(10,5))

bars = plt.bar(
    dish_count.index,
    dish_count.values,
    color=['red','blue','green','orange','purple','pink','cyan','gold']
)

plt.title("Most Popular Korean Dishes")
plt.xlabel("Food Item")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)

# Add values on bars
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             y + 0.05,
             int(y),
             ha='center')

plt.show()
```

# High order locations VS Low order locations

```python
import pandas as pd
import matplotlib.pyplot as plt

# Location Order Count
location_count = df['Location'].value_counts().reset_index()
location_count.columns = ['Location', 'Orders']

print(location_count)

# Heat Map
heat_data = location_count.set_index('Location')

plt.figure(figsize=(6,4))
plt.imshow(heat_data[['Orders']],
           cmap='YlOrRd',
           aspect='auto')

plt.colorbar(label='Orders')
plt.yticks(range(len(heat_data.index)), heat_data.index)
plt.xticks([0], ['Order Density'])

plt.title('High & Low Order Locations Heat Map')
plt.show()
```

# Dashboard for Business Insights

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
data = {
    'Customer_Name': ['Min-Jun','Ji-Hoon','Seo-Yeon','Ha-Neul','Yu-Jin',
                      'Min-Jun','Ji-Hoon','Seo-Yeon','Ha-Neul','Yu-Jin',
                      'Jae-Hyun','Su-Min','Dong-Hyun','Eun-Ji','Hye-Jin'],

    'Food_Item': ['Kimchi Fried Rice','Bibimbap','Tteokbokki','Bulgogi','Japchae',
                  'Kimchi Fried Rice','Bibimbap','Tteokbokki','Bulgogi','Japchae',
                  'Samgyeopsal','Kimbap','Jajangmyeon','Kimchi Fried Rice','Bibimbap'],

    'Order_Time': ['09:00','10:00','12:00','13:00','18:00',
                   '19:00','20:00','21:00','18:00','19:00',
                   '20:00','12:00','13:00','18:00','19:00'],

    'Location': ['Gangnam','Myeongdong','Hongdae','Itaewon','Gangnam',
                 'Hongdae','Gangnam','Myeongdong','Itaewon','Hongdae',
                 'Gangnam','Myeongdong','Hongdae','Gangnam','Itaewon'],

    'Price': [250,300,200,450,350,250,300,200,450,350,500,150,280,250,300]
}

df = pd.DataFrame(data)

# Analysis
peak_orders = df.groupby('Order_Time').size()
dish_count = df['Food_Item'].value_counts()
location_count = df['Location'].value_counts()

# Dashboard Layout
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 1. Peak Order Times
axes[0,0].plot(peak_orders.index,
               peak_orders.values,
               marker='o',
               linewidth=3)
axes[0,0].set_title('Peak Order Times')
axes[0,0].set_xlabel('Order Time')
axes[0,0].set_ylabel('Orders')
axes[0,0].grid(True)

# 2. Most Popular Dishes
axes[0,1].bar(dish_count.index,
              dish_count.values)
axes[0,1].set_title('Most Popular Dishes')
axes[0,1].set_xlabel('Food Item')
axes[0,1].set_ylabel('Orders')
axes[0,1].tick_params(axis='x', rotation=45)

# 3. Location Heat Map
heat_data = location_count.values.reshape(-1,1)

im = axes[1,0].imshow(
    heat_data,
    cmap='YlOrRd',
    aspect='auto'
)

axes[1,0].set_title('Location Order Density')
axes[1,0].set_yticks(range(len(location_count.index)))
axes[1,0].set_yticklabels(location_count.index)
axes[1,0].set_xticks([0])
axes[1,0].set_xticklabels(['Orders'])

fig.colorbar(im, ax=axes[1,0])

# 4. Summary Box
axes[1,1].axis('off')

summary = f"""
Total Orders : {len(df)}

Total Revenue : ₹{df['Price'].sum()}

Average Order Value : ₹{round(df['Price'].mean(),2)}

Top Dish :
{dish_count.idxmax()}

Top Location :
{location_count.idxmax()}
"""

axes[1,1].text(
    0.05,
    0.5,
    summary,
    fontsize=12,
    bbox=dict(facecolor='lightyellow')
)

plt.suptitle(
    'Food Delivery Sales & Customer Insights Dashboard',
    fontsize=18,
    fontweight='bold'
)

plt.tight_layout()
plt.show()
```
